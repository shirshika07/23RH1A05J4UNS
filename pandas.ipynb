```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
df=pd.read_csv("/content/Mall_Customers (1).csv")
df
```

```python
plt.figure(figsize=(10,5))
plt.title("Before Clustering")
sns.scatterplot(x=X['Annual Income (k$)'],y=X['Spending Score (1-100)'],s=100)
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.grid(True)
plt.show()
```

```python
kmeans=KMeans(n_clusters=8,random_state=42)
df['cluster']=kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(10,5))
plt.title("After Clustering")
sns.scatterplot(data=df,x='Annual Income (k$)',y='Spending Score (1-100)',hue='cluster',palette='Set2',s=100)
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.grid(True)
plt.legend("cluster")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
df=pd.read_csv("/content/Mall_Customers (1).csv")
df
```

```python
X=df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit_transform(X)
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
linked=linkage(X_scaler,method='ward')
plt.figure(figsize=(12,6))
dendrogram(linked,orientation='top',distance_sort='descending',show_leaf_counts=True)
plt.title("Before Clustering")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.grid(True)
plt.show()
```

```python
hier_cluster=AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward')
df['cluster']=hier_cluster.fit_predict(X_scaler)
```

```python
plt.figure(figsize=(10,5))
plt.title("After Clustering")
sns.scatterplot(data=df,x='Annual Income (k$)',y='Spending Score (1-100)',hue='cluster',palette='Set2',s=100)
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.grid(True)
plt.legend("cluster")
plt.show()
```
